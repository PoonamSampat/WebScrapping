```python
from bs4 import BeautifulSoup
```

```python
import requests   # use this only, not urllibx anymore.
page = requests.get("https://www.tripadvisor.in/Hotels-g303877-Panjim_North_Goa_District_Goa-Hotels.html").text
soup_obj = BeautifulSoup(page, "lxml")
```

```python
source_site = soup_obj.find_all('div', class_ = "listing")
```

```python
constant_URL = "https://www.tripadvisor.in"
links = []

for each_hotel in source_site:
    hotel_url = constant_URL + each_hotel.find('a').attrs['href']
    links.append(hotel_url)
    
links = links[:10]
```

```python
hotels_list = []

for each_hotel in source_site:
    get_name = each_hotel.find('div',class_ = 'listing_title')
    hotels_list.append(get_name.a.text)

hotels_list = hotels_list[:10]
```

```python
current_price = []

def find_price_on_page():
    page_detail = requests.get(hotel_page_url).text
    soup_obj_detail = BeautifulSoup(page_detail, "lxml")
    price_tag = soup_obj_detail.find('div', class_ = "price ")
    if price_tag is not None:
        price_tag = price_tag.text
        price_tag = price_tag.replace(u'\u20b9\xa0', '')
        current_price.append(price_tag)
    else:
        price_tag="Price Not found"
        current_price.append(price_tag)    

for each_hotel in source_site:
    hotel_page_url = constant_URL + each_hotel.find('a').attrs['href']
    price_tag = each_hotel.find('div', class_ = "price")   
    if price_tag is not None:
        price_tag = price_tag.text
        price_tag = price_tag.replace(u'\u20b9\xa0', '')
        current_price.append(price_tag)
    else:
        find_price_on_page()
    
current_price = current_price[:10]
```

```python
Deal_savings = []

def find_savings_on_page():
    page_detail = requests.get(hotel_page_url).text
    soup_obj_detail = BeautifulSoup(page_detail, "lxml")
    savings_tag = soup_obj_detail.find('span', class_ = "saveMoney")
    if savings_tag is not None:
        savings_tag = savings_tag.text
        savings_tag = savings_tag.replace(u'\u20b9\xa0', '')
        Deal_savings.append(savings_tag)
    else:
        savings_tag="No Savings on deal"
        Deal_savings.append(savings_tag)
    
for each_hotel in source_site:
    hotel_page_url = constant_URL + each_hotel.find('a').attrs['href']
    savings_tag = each_hotel.find('div', class_ = "save")   
    if savings_tag is not None:
        savings_tag = savings_tag.text
        savings_tag = savings_tag.replace(u'\u20b9\xa0', '')
        Deal_savings.append(savings_tag)
    else:
        find_savings_on_page()
    
Deal_savings = Deal_savings[:10]
```

```python
Facilities_here = []

for each_hotel in links:
    Hotel_has = []
    page_detail = requests.get(each_hotel).text
    soup_obj_detail = BeautifulSoup(page_detail, "lxml")
    
    Facilities = soup_obj_detail.find_all('div', class_ = "highlightedAmenity")
    
    if Facilities is not None and Facilities != []:
        for f in Facilities:
            Hotel_has.append(f.text)
        Facilities_here.append(Hotel_has)
    else:
        Facilities = "No Facilities"
        Facilities_here.append(Facilities)
               
Facilities_here = Facilities_here[:10]
```

```python
address = [] #Getting address of each hotel & storing in this list

for hotel_details in links:
    page_detail = requests.get(hotel_details).text #Run the links we fetched for each hotel, individual review page    
    soup_obj_detail = BeautifulSoup(page_detail, "lxml") #Get data from page of every Hotel
    
    hotel_details_page_streetaddress = soup_obj_detail.find('span', class_ = "street-address")
    if hotel_details_page_streetaddress is not None:
        hotel_details_page_streetaddress = hotel_details_page_streetaddress.text
    else:
        hotel_details_page_streetaddress="Not found on Hotel review page"
        
    hotel_details_page_locality = soup_obj_detail.find('span', class_ = "locality")
    if hotel_details_page_locality is not None:
        hotel_details_page_locality = hotel_details_page_locality.text
    else:
        hotel_details_page_locality="Not found on Hotel review page"
    
    hotel_details_page_country_name = soup_obj_detail.find('span', class_ = "country-name")
    if hotel_details_page_country_name is not None:
        hotel_details_page_country_name = hotel_details_page_country_name.text
    else:
        hotel_details_page_country_name="Not found on Hotel review page"   
        
    address.append(hotel_details_page_streetaddress +','+hotel_details_page_locality+hotel_details_page_country_name)
        
address =address[:10]   
```

```python
rating_main = soup_obj.find_all('div', class_ = "prw_rup prw_common_responsive_rating_and_review_count linespace is-shown-at-mobile")
 
const_str = " of 5 bubbles"    
ratings_list = []    
for rating_rows in rating_main:
    complete_ratings = rating_rows.find('span').attrs['alt']
    complete_ratings = complete_ratings.replace(const_str,'')
    ratings_list.append(complete_ratings)
    
ratings_list = ratings_list[:10]
```

```python
 #For each hotel, scrape these details (hotel name, price, tripadvisor url, savings,  
 #number_of_reviews, average rating, facilities such as wifi, parking, etc) 
    
reviews =soup_obj.find_all('a', class_ = "review_count") 

reviews_list = [hotels.string for hotels in reviews]
reviews_list  = reviews_list[:10]
```

```python
values_rating =soup_obj.find_all('div', class_ = "popindex")    
values_rating

values_rating_list = [hotels.string for hotels in values_rating]
values_rating_list = values_rating_list[:10]
```

```python
import pandas as pd
TripAdvisor = pd.DataFrame(
{'A)Hotels': hotels_list,
 'B)Ratings':ratings_list,
 'C)Best Offer(INR)': current_price,
 'D)Savings(INR)': Deal_savings,
 'E)Facilities': Facilities_here,
 'F)Address': address,
 'G)Reviews': reviews_list,
 'H)Rating Desc': values_rating_list,
 'I)deeplinks':links,
})

TripAdvisor
```
